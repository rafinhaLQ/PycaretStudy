# Regressão

```python
from pycaret.regression import *
```

```python
s = setup(data, target='Risk', categorical_features=['isMale', 'isBlack', 'isSmoker', 'isDiabetic', 'isHypertensive'])
```

```python
best = compare_models()
```

```python
plot_model(best, plot='residuals')
```

```python
plot_model(best, plot='error')
```

```python
plot_model(best, plot='feature')
```
